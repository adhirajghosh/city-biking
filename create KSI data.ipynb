## Processing of the Data:

### Utils

```python
def make_array(string, length):
    return np.array(string.split(" ")).reshape((-1, length))

def unify_into_df(burrow_strings, columns):
    return pd.DataFrame([
        *make_array(burrow_strings[0], len(columns)),
        *make_array(burrow_strings[1], len(columns)),
        *make_array(burrow_strings[2], len(columns)),
        *make_array(burrow_strings[3], len(columns)),
        *make_array(burrow_strings[4], len(columns))],
        columns=columns)
```

### Bicycle-Motor-Incidents

```python
columns = ['borough',
           'precinct',
           'crashes',
           'bicyclist injuries',
           'bicyclist fatalities',
           'motor vehicle occupant injuries',
           'motor vehicle occupant fatalities'
           ]

Table_14_BM = unify_into_df([
    table14_Manhattan_BM,
    table14_Bronx_BM,
    table14_Brooklyn_BM,
    table14_Queens_BM,
    table14_StatenIsland_BM
    ], columns)

Table_15_BM = unify_into_df([
    table15_Manhattan_BM,
    table15_Bronx_BM,
    table15_Brooklyn_BM,
    table15_Queens_BM,
    table15_StatenIsland_BM
    ], columns)

Table_16_BM = unify_into_df([
    table16_Manhattan_BM,
    table16_Bronx_BM,
    table16_Brooklyn_BM,
    table16_Queens_BM,
    table16_StatenIsland_BM
    ], columns)

Table_14_BM.to_csv('data/nyc/KSI14_BM.csv', encoding='utf-8', index=False)
Table_15_BM.to_csv('data/nyc/KSI15_BM.csv', encoding='utf-8', index=False)
Table_16_BM.to_csv('data/nyc/KSI16_BM.csv', encoding='utf-8', index=False)
```
