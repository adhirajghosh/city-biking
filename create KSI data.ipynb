```python
def make_array(string):
    return np.array(string.split(" ")).reshape((-1, 7))

columns = ['burrow',
           'precinct',
           'crashes',
           'Bicyclist Injuries',
           'Bicyclist Fatalities',
           'Motor Vehicle Occupant Injuries',
           'Motor Vehicle Occupant Fatalities'
           ]


def unify_into_df(burrow_strings):
    return pd.DataFrame([
        *make_array(burrow_strings[0]),
        *make_array(burrow_strings[1]),
        *make_array(burrow_strings[2]),
        *make_array(burrow_strings[3]),
        *make_array(burrow_strings[4])],
        columns=columns)


Table_14_BM = unify_into_df([
    table14_Manhattan_BM,
    table14_Bronx_BM,
    table14_Brooklyn_BM,
    table14_Queens_BM,
    table14_StatenIsland_BM
])
Table_15_BM = unify_into_df([
    table15_Manhattan_BM,
    table15_Bronx_BM,
    table15_Brooklyn_BM,
    table15_Queens_BM,
    table15_StatenIsland_BM
])
Table_16_BM = unify_into_df([
    table16_Manhattan_BM,
    table16_Bronx_BM,
    table16_Brooklyn_BM,
    table16_Queens_BM,
    table16_StatenIsland_BM
])
```
